```python
def addEdge(adj, v, w):     
    adj[v].append(w)
    adj[w].append(v) 
    
    return adj
```

```python
def greedyColoring(adj, V):
     
    result = [-1] * V
 
    
    result[0] = 0;

    available = [False] * V

    for u in range(1, V):

        for i in adj[u]:
            if (result[i] != -1):
                available[result[i]] = True
 

        cr = 0
        while cr < V:
            if (available[cr] == False):
                break
             
            cr += 1

        result[u] = cr

        for i in adj[u]:
            if (result[i] != -1):
                available[result[i]] = False

    for u in range(V):
        print("Vertex", u, " --->  Color", result[u])
    
    return result
```
